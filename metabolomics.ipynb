```python
import requests

def get_cid_for_compound(compound_name):
    base_url = 'https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/name/'
    url = base_url + compound_name + '/cids/TXT'
    response = requests.get(url)

    if response.status_code == 200:
        cid = response.text.strip()
        return cid
    else:
        print(f"Failed to retrieve CID for {compound_name}. Status code: {response.status_code}")
        return None

# Example usage
compound_names = ['1-Monopalmitin, 2TMS derivative']
for name in compound_names:
    cid = get_cid_for_compound(name)
    if cid:
        print(f"Compound: {name}, CID: {cid}")
```

```python
from umap import UMAP

X = df_bn[['wtr', 'oer']]
umap = UMAP(n_components=2)
X_umap = umap.fit_transform(X)
plt.figure(figsize=(8, 6))
plt.scatter(X_umap[:, 0], X_umap[:, 1], c='blue', s=5)
for i, txt in enumerate(df_bn.index):
    plt.annotate(txt, (X_umap[i, 0], X_umap[i, 1]), fontsize=7)
plt.xlabel('UMAP Component 1')
plt.ylabel('UMAP Component 2')
plt.title('UMAP Visualization')
plt.show()
```

UMAP tends to preserve the local structure of the data, meaning that nearby points in the high-dimensional space are likely to be close to each other in the UMAP visualization. Therefore, clusters of points in the UMAP plot may represent clusters or groups of similar data points in the original high-dimensional space.
